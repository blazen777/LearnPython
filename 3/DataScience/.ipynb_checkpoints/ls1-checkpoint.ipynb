```python
import csv
```

```python
streets = {}
```

```python
with open('ostanovka.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f,delimiter=';')
    for row in reader:
        if(row['Street']) in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
        if(row['Street'] == 'проезд без названия'):
            print(f'Название:{row["Name"]} {row["District"]}')
```

```python
sorted(streets.items(), key=lambda item: item[1], reverse=True)
```
